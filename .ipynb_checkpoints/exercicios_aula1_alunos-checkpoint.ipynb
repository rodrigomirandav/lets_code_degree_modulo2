1. Crie uma classe Bola cujos atributos são cor e raio. Crie um método que imprime a cor da bola. Crie um método para calcular a área dessa bola. Crie um método para calcular o volume da bola. Crie um objeto dessa classe e calcule a área e o volume, imprimindo ambos em seguida. 
    Obs.:

    Área da esfera = 4*3.14*r*r;

    Volume da esfera = 4*3.14*r*r*r/3

```python
import math

class Bola:
    
    def __init__(self, cor, raio):
        self.cor = cor
        self.raio = raio
        
    def area(self):
        return (4 * math.pi * (self.raio ** 2))
    
    def volume(self):
        return (  (4 / 3)  *  (math.pi * (self.raio ** 3)) )
    
    def imprimir_cor(self):
        print(f"Cor da bola => {self.cor}")
        
    def __repr__(self):
        return f"Bola de raio {self.raio} e cor {self.cor}"
    
bola1 = Bola("Amarela", 1)
bola1.imprimir_cor()
print(bola1.area())
print(bola1.volume())
bola1
```

```python
class Retangulo:
    
    def __init__(self, lado_a, lado_b):
        self.lado_a = lado_a
        self.lado_b = lado_b
        
    def area(self):
        return self.lado_a * self.lado_b
    
    def __repr__(self):
        return f"Retangulo de lado a = {self.lado_a} e lado b = {self.lado_b}"
    
retangulo1 = Retangulo(25,4)
print(retangulo1.area())
retangulo1
```

3. Crie uma classe Funcionario cujos atributos são nome e e-mail. Guarde as horas trabalhadas em um dicionário cujas chaves são o mês em questão e, em outro dicionário, guarde o salário por hora relativo ao mês em questão. Crie um método que retorna o salário mensal do funcionário.

```python
class Funcionario:
    
    def __init__(self,nome, email):
        self.nome = nome
        self.email = email
        
        self.horas_trabalhadas = {}
        self.salario_hora = {}
        
    def cadastra_hora_trabalhada_mes(self, mes, hora):
        self.horas_trabalhadas[mes.lower()] = hora
        
    def cadasta_salario_hora_mes(self, mes, salario_hora):
        self.salario_hora[mes.lower()] = salario_hora
        
    def salario_mensal(self,mes):
        mes = mes.lower()
        if ( (mes in self.horas_trabalhadas) and (mes in self.salario_hora) ):
            return self.horas_trabalhadas[mes] * self.salario_hora[mes]
        else:
            return 0
        
    def __repr__(self):
        string_retorno = f"""Nome: {self.nome}\nEmail : {self.email}\n"""
    
        for mes in (self.horas_trabalhadas):
            string_retorno += f"""Mês {mes}:\nHora mês -> {self.horas_trabalhadas[mes]}\nValor hora mês -> {self.salario_hora[mes]}
                            """
        return string_retorno
    
funcionario_1 = Funcionario("Rodrigo", "rodrigomirandav250@gmail.com")
funcionario_1.cadastra_hora_trabalhada_mes("outubro", 30)
funcionario_1.cadasta_salario_hora_mes("outubro", 65.0)
funcionario_1.salario_mensal("outubro")
funcionario_1
```

4. Crie uma classe Televisor cujos atributos são:

    a. fabricante;

    b. modelo;

    c. canal atual;

    d. lista de canais; e

    e. volume.

    Faça métodos para aumentar/diminuir volume, trocar o canal e sintonizar um novo canal, que adiciona um novo canal à lista de canais (somente se esse canal não estiver nessa lista). No atributo lista de canais, devem estar armazenados todos os canais já sintonizados dessa TV.

    Obs.: O volume não pode ser menor que zero e maior que cem; só se pode trocar para um canal que já esteja na lista de canais.

```python
class Televisor:
    
    def __init__(self, fabricante, modelo, lista_canais=[]):
        self.fabricante = fabricante
        self.modelo = modelo
        self.lista_canais = lista_canais
        self.canal = None
        self.volume = 0
        
    def aumentar_volume(self, volume):
        volume = self.volume + volume
        
        if volume > 100:
            self.volume = 100
        else:
            self.volume = volume
        
    def diminuir_volume(self, volume):
        volume = self.volume - volume
        
        if volume < 0:
            self.volume = 0
        else:
            self.volume = volume
            
    def mostrar_volume(self):
        print(f"Volume => {self.volume}")
        
    def trocar_canal(self, canal):
        if canal in self.lista_canais:
            self.canal = canal
    
    def sintonizar_canal(self, canal):
        if canal not in self.lista_canais:
            self.lista_canais.append(canal)
    
    def mostrar_canal(self):
        print(f"Canal => {self.canal}")
        
    def mostrar_canais_sintonizados(self):
        print(f"Canais => {self.lista_canais}")
        
    def __repr__(self):
        return f"""
        Fabricante: {self.fabricante}
        Modelo: {self.modelo}
        Lista canais: {self.lista_canais}
        Canal: {self.canal}
        Volume: {self.volume}
        """
    
televisor1 = Televisor("Samsung", "SXP-15856")
televisor1.aumentar_volume(5)
televisor1.mostrar_volume()
televisor1.diminuir_volume(10)
televisor1.mostrar_volume()
televisor1.trocar_canal(4)
televisor1.sintonizar_canal(10)
televisor1.mostrar_canal()
televisor1.mostrar_canais_sintonizados()
televisor1.trocar_canal(10)
televisor1.mostrar_canal()
televisor1.sintonizar_canal(5)
televisor1.mostrar_canais_sintonizados()
televisor1.trocar_canal(5)
televisor1.mostrar_canal()
televisor1.aumentar_volume(200)
televisor1.mostrar_volume()
televisor1
```

6. O módulo time possui a função time.sleep(x), que faz seu programa “dormir” por x segundos. Utilizando essa função, crie uma classe Cronômetro e faça um programa que cronometre o tempo.

```python
import time
from IPython.display import clear_output

class Cronometro:
    '''
    Cria uma representação de um cronometro 
    
    Atributos
    ----------
    dias : int
        Dias passados
    horas : int
        Horas passadas
    minutos : int
        Minutos passados
    segundos : int
        Segundos passados
    '''
    
    def __init__(self):
        '''
        Construtor
        '''

        # Criando os atributos da classe Cronometro
        self.dias = 0
        self.horas = 0
        self.minutos = 0
        self.segundos = 0
        
    def iniciar_cronometro(self):
        '''
        função para inicializar o cronometro       
        '''
        while True:
            time.sleep(1)
            self.segundos += 1

            if self.segundos == 60:
                self.segundos = 0
                self.minutos += 1

            if self.minutos == 60:
                self.minutos = 0
                self.horas += 1

            if self.horas == 24:
                self.horas = 0
                self.dias += 1

            self.imprmir_tempo()
        
    def imprmir_tempo(self):
        '''
            Função que imprime o tempo     
        '''
        clear_output(wait=True)
        if self.dias == 0:
            print(f'{self.horas:02d}:{self.minutos:02d}:{self.segundos:02d}')
        else:
            print(f'{self.dias} dia(s) {self.horas:02d}:{self.minutos:02d}:{self.segundos:02d}')
            
    def __repr__(self):
        if self.dias == 0:
            return f'{self.horas:02d}:{self.minutos:02d}:{self.segundos:02d}'
        else:
            return f'{self.dias} dia(s) {self.horas:02d}:{self.minutos:02d}:{self.segundos:02d}'
    
cronometro1 = Cronometro()
cronometro1.iniciar_cronometro()
cronometro1
```

7. Crie uma modelagem de classes para uma agenda capaz de armazenar contatos. Através dessa agenda é possível incluir, remover, buscar e listar contatos já cadastrados.

```python
class Contato:
    '''
    Cria uma representação de um contato 
    
    Atributos
    ----------
    nome : str
        Nome do contato
    telefone : str
        Telefone do contato
    '''
    
    def __init__(self, nome, telefone):
        '''
        Construtor

        Parâmetros
        ----------
        nome : str
            Recebe um nome como parametro
        telefone : str
            Recebe um telefone como parametro
        '''

        # Criando os atributos da classe Contato
        self.nome = nome
        self.telefone = telefone


class Agenda:
    '''
    Cria uma representação de uma Agenda
    
    Atributos
    ----------
    contatos : list:`Contato`
        Guarda um contato na agenda
    '''
    
    def __init__(self):
        '''
        Construtor
        '''
        self.contatos = []
        
    def adicionar_contato(self, contato):
        '''
        Adiciona um contato na agenda

        Parâmetros
        ----------
        contato : Contato
            Recebe um contato para ser adicionado
        
        Retornos
        -------
        bool
            Verdadeiro se foi adicionado ou falso caso já esteja 
            na lista de contatos
        '''
        if contato not in self.contatos:
            self.contatos.append(contato)
            return True
        return False
    
    def imprime_contato(self, contato):
        '''
        Imprime um contato

        Parâmetros
        ----------
        contato : Contato
            Recebe um contato para ser impresso
        '''
        print(f"Nome => {contato.nome}\nTelefone=> {contato.telefone}\n")
    
    def listar_contatos(self):
        '''
        Lista todos os contatos da agenda
        '''
        for contato in self.contatos:
            self.imprime_contato(contato)
            
            
    def buscar_por_nome(self, texto_busca):
        '''
        Busca um contato através do nome, ele retorna todos os contatos
        onde possui parte do texto buscado

        Parâmetros
        ----------
        texto_busca : str
            Recebe um texto de busca para ser encontrado no nome dos contatos
        
        Retornos
        -------
        list:`Contato`
            Retorna uma lista de contatos
        '''
        return [contato for contato in self.contatos if texto_busca.lower() in contato.nome.lower()]
```

10) Crie uma classe ContaCorrente com os atributos cliente (que deve ser um objeto da classe Cliente) e saldo. Crie métodos para depósito, saque e transferência. Os métodos de saque e transferência devem verificar se é possível realizar a transação.

```python
class Cliente:
    '''
    Cria uma representação de um cliente 
    
    Atributos
    ----------
    nome : str
        Nome do cliente
    idade : int
        Idade do cliente
    email: str
        Email do cliente
    '''
    
    def __init__(self, nome, idade, email):
        '''
        Construtor

        Parâmetros
        ----------
        nome : str
            Recebe um nome como parametro
        idade : int
            Recebe uma idade como parametro
        email : str
            Recebe um email como parametro
        '''

        # Criando os atributos da classe Cliente
        self.nome = nome
        self.idade = idade
        self.email = email
        
    def imprimir_cliente(self):
        '''
        Função que imprime os dados do cliente     
        '''
        print(f"""
            Nome: {self.nome}\n
            Idade: {self.idade}\n
            E-mail: {self.email}
        """)
        
class ContaCorrente():
    '''
    Cria uma representação de uma conta 
    
    Atributos
    ----------
    cliente : Cliente
        Guarda o cliente da conta
    saldo : float
        Saldo da conta corrente
    '''
    
    def __init__(self, cliente):
        '''
        Construtor

        Parâmetros
        ----------
        cliente : Cliente
            Recebe um cliente tornando proprietário da conta
        '''
        if(not isinstance(cliente, Cliente)):
            raise ValueError('É necessário passar um cliente para a conta')
            
        self.cliente = cliente
        self.saldo = 0.0
        
    def __repr__(self):
        return f"""
        Cliente: {self.cliente.nome}
        Valor: {self.saldo:.2f}
        """
        
    def depositar(self, valor):
        '''
        Deposita um valor passado como parâmetro na conta do usuário,
        alterando o saldo da conta.
        Caso o parametro não seja passado corretamente o sistema levanta uma 
        exceção

        Parâmetros
        ----------
        valor : float
            Recebe o valor a ser depositado
        
        Retornos
        -------
        bool:
            Retorna se a operação foi realizada com sucesso
        '''
        if(not (isinstance(valor, float) or isinstance(valor, int))):
            raise ValueError('O valor precisa ser númerico')
            
        if(valor <= 0):
            raise ValueError('O valor precisa ser maior que zero')
            
        self.saldo += valor
        return True
    
    def sacar(self, valor):
        '''
        Saca um valor passado como parâmetro da conta do usuário,
        alterando o saldo da conta.
        Caso o parametro não seja passado corretamente, 
        ou caso não for possível realizar o saque por falta de saldo.
        O sistema levanta uma exceção

        Parâmetros
        ----------
        valor : float
            Recebe o valor a ser sacado
        
        Retornos
        -------
        bool:
            Retorna se a operação foi realizada com sucesso
        '''
        if(not (isinstance(valor, float) or isinstance(valor, int))):
            raise ValueError('O valor precisa ser númerico')
            
        if(valor <= 0):
            raise ValueError('O valor precisa ser maior que zero')
            
        if( (self.saldo - valor) < 0):
            raise ValueError('A conta não possui saldo suficiente para esta operação')
            
        self.saldo -= valor
        return True
    
    def transferir(self, conta_destino, valor):
        '''
        Transfere o valor solicitado para a conta de destino.
        Caso o parametro não seja passado corretamente, 
        ou caso não for possível realizar a transferência por falta de saldo.
        O sistema levanta uma exceção

        Parâmetros
        ----------
        conta_destino: ContaCorrente
            Recebe uma conta corrente para efetuar entrada do valor de trasnferência
        valor : float
            Recebe o valor a ser transferido

        Retornos
        -------
        bool:
            Retorna se a operação foi realizada com sucesso
        '''
        if(not isinstance(conta_destino, ContaCorrente)):
            raise ValueError('É necessário passar uma conta corrente para transferir')
            
        try:
            self.sacar(valor)
        except Exception as erro:
            print(erro)
            return False      

        conta_destino.depositar(valor)
        return True
    
cliente1 = Cliente("Rodrigo", 33, "")
conta1 = ContaCorrente(cliente1)
conta1.depositar(200)
print(conta1)
conta1.depositar(200)
print(conta1)
conta1.sacar(200)
print(conta1)
# Erro de saldo
#conta1.sacar(300)
print(conta1)

cliente2 = Cliente("Maria", 20, "")
conta2 = ContaCorrente(cliente2)
print(conta2)

conta1.transferir(conta2, 100)
print(conta1)
print(conta2)

conta1.transferir(conta2, 100)
print(conta1)
print(conta2)

conta1.transferir(conta2, 100)
print(conta1)
print(conta2)

conta2.transferir(conta1, 150.99)
print(conta1)
print(conta2)
```

11) Crie uma classe Fração cujos atributos são numerador (número de cima) e denominador (número de baixo).

Implemente os métodos de adição, subtração, multiplicação, divisão que retornam objetos do tipo Fração.

Implemente também o método _ repr _.

Implemente métodos para comparação: igualdade (==) e desigualdades (!=, <=, >=, < e >).

```python
class Fracao:
    
    def __init__(self, numerador, denominador):
        if (denominador == 0):
            raise ValueError("Denominador não pode ser 0")
            
        self.numerador = numerador
        self.denominador = denominador
        
    def __repr__(self):
        return f"{self.numerador}/{self.denominador}"

    def valor(self):
        return self.numerador/self.denominador
        
    def __add__(self, other):
        num = self.numerador*other.denominador +\
        self.denominador*other.numerador  
        
        den = self.denominador*other.denominador
        
        return Fracao(num, den)
    
    def __sub__(self, other):
        num = self.numerador*other.denominador -\
        self.denominador*other.numerador  
        
        den = self.denominador*other.denominador
        
        return Fracao(num, den)
    
    def __mul__(self, other):
        num = self.numerador*other.numerador        
        den = self.denominador*other.denominador
        
        return Fracao(num,den)
    
    def __truediv__(self, other):
        num = self.numerador*other.denominador
        
        den = self.denominador*other.numerador
        
        return Fracao(num, den)
        
    def __eq__(self,other):
        return self.valor() == other.valor()
    
    def __lt__(self, other):
        return self.valor() < other.valor()
    
    def __le__(self, other):
        return self.valor() <= other.valor()
    
    def __gt__(self, other):
        return self.valor() > other.valor()
    
    def __ge__(self, other):
        return self.valor() >= other.valor()
```

12) Crie uma classe Data cujos atributos são dia, mês e ano. Implemente métodos _ repr _ e para comparação: igualdade (==) e desigualdades (!=, <=, >=, < e >).

```python
class Data:
    '''
    Cria uma representação de uma data 
    
    Atributos
    ----------
    dia : int
        Guarda o dia da data
    mes : int
        Guarda o mês da data
    ano : int
        Guarda o ano da data
    '''
    
    def __init__(self, dia, mes, ano):
        '''
        Construtor

        Parâmetros
        ----------
        dia : int
            Recebe o dia
        mes : int
            Recebe o mês
        ano : int
            Recebe o ano
        '''
        if(not isinstance(dia, int)):
            raise ValueError('O dia precisa ser inteiro')
            
        if(not isinstance(mes, int)):
            raise ValueError('O mês precisa ser inteiro')
            
        if(not isinstance(ano, int)):
            raise ValueError('O ano precisa ser inteiro')
            
        if(not dia in range(1,32)):
            raise ValueError('O dia precisa estar compreendido entre 1 a 31')
            
        if(not mes in range(1,13)):
            raise ValueError('O mês precisa estar compreendido entre 1 a 12')
            
        self.dia = dia
        self.mes = mes
        self.ano = ano
        
    def __repr__(self):
        return f'{self.dia:02d}/{self.mes:02d}/{self.ano:04d}'
    
    def __eq__(self, other):
        if self.ano != other.ano:
            return False
        
        if self.mes != other.mes:
            return False
        
        if self.dia != other.dia:
            return False
        
        return True
    
    def __ne__(self, other):
        return not self.__eq__(other)
    
    def __lt__(self, other):
        if self.__eq__(other):
            return False       
        
        if self.ano > other.ano:
            return False        
        
        if self.mes > other.mes:
            return False
        
        if self.dia > other.dia:
            return False
        
        return True
    
    def __le__(self, other):
        return (self.__eq__(other) or self.__lt__(other))
    
    def __gt__(self, other):
        if self.__eq__(other):
            return False       
        
        if self.ano < other.ano:
            return False        
        
        if self.mes < other.mes:
            return False
        
        if self.dia < other.dia:
            return False
        
        return True
    
    def __ge__(self, other):
        return (self.__eq__(other) or self.__gt__(other))
    
dia1 = Data(4,1,1987)
dia1
dia2 = Data(3,1,1987)
#erro
#dia2 = Data(31,13,1987)
#dia2 = Data(32,13,1987)
print(dia1.__eq__(dia2))
print(dia1.__ne__(dia2))
print(dia1.__lt__(dia2))
print(dia1.__le__(dia2))
print(dia1.__gt__(dia2))
print(dia1.__ge__(dia2))
```

14) Faça uma classe ContaVip que difere da ContaCorrente por ter cheque especial (novo atributo) e é filha da classe ContaCorrente. Você precisa implementar os métodos para saque, transferência ou depósito?

```python
class Cliente:
    '''
    Cria uma representação de um cliente 
    
    Atributos
    ----------
    nome : str
        Nome do cliente
    idade : int
        Idade do cliente
    email: str
        Email do cliente
    '''
    
    def __init__(self, nome, idade, email):
        '''
        Construtor

        Parâmetros
        ----------
        nome : str
            Recebe um nome como parametro
        idade : int
            Recebe uma idade como parametro
        email : str
            Recebe um email como parametro
        '''

        # Criando os atributos da classe Cliente
        self.nome = nome
        self.idade = idade
        self.email = email
        
    def imprimir_cliente(self):
        '''
        Função que imprime os dados do cliente     
        '''
        print(f"""
            Nome: {self.nome}\n
            Idade: {self.idade}\n
            E-mail: {self.email}
        """)
        
class ContaCorrente():
    '''
    Cria uma representação de uma conta 
    
    Atributos
    ----------
    cliente : Cliente
        Guarda o cliente da conta
    saldo : float
        Saldo da conta corrente
    '''
    
    def __init__(self, cliente):
        '''
        Construtor

        Parâmetros
        ----------
        cliente : Cliente
            Recebe um cliente tornando proprietário da conta
        '''
        if(not isinstance(cliente, Cliente)):
            raise ValueError('É necessário passar um cliente para a conta')
            
        self.cliente = cliente
        self.saldo = 0.0
        
    def __repr__(self):
        return f"""
        Cliente: {self.cliente.nome}
        Valor: {self.saldo:.2f}
        """
        
    def depositar(self, valor):
        '''
        Deposita um valor passado como parâmetro na conta do usuário,
        alterando o saldo da conta.
        Caso o parametro não seja passado corretamente o sistema levanta uma 
        exceção

        Parâmetros
        ----------
        valor : float
            Recebe o valor a ser depositado
        
        Retornos
        -------
        bool:
            Retorna se a operação foi realizada com sucesso
        '''
        if(not (isinstance(valor, float) or isinstance(valor, int))):
            raise ValueError('O valor precisa ser númerico')
            
        if(valor <= 0):
            raise ValueError('O valor precisa ser maior que zero')
            
        self.saldo += valor
        return True
    
    def sacar(self, valor):
        '''
        Saca um valor passado como parâmetro da conta do usuário,
        alterando o saldo da conta.
        Caso o parametro não seja passado corretamente, 
        ou caso não for possível realizar o saque por falta de saldo.
        O sistema levanta uma exceção

        Parâmetros
        ----------
        valor : float
            Recebe o valor a ser sacado
        
        Retornos
        -------
        bool:
            Retorna se a operação foi realizada com sucesso
        '''
        if(not (isinstance(valor, float) or isinstance(valor, int))):
            raise ValueError('O valor precisa ser númerico')
            
        if(valor <= 0):
            raise ValueError('O valor precisa ser maior que zero')
            
        if( (self.saldo - valor) < 0):
            raise ValueError('A conta não possui saldo suficiente para esta operação')
            
        self.saldo -= valor
        return True
    
    def transferir(self, conta_destino, valor):
        '''
        Transfere o valor solicitado para a conta de destino.
        Caso o parametro não seja passado corretamente, 
        ou caso não for possível realizar a transferência por falta de saldo.
        O sistema levanta uma exceção

        Parâmetros
        ----------
        conta_destino: ContaCorrente
            Recebe uma conta corrente para efetuar entrada do valor de trasnferência
        valor : float
            Recebe o valor a ser transferido

        Retornos
        -------
        bool:
            Retorna se a operação foi realizada com sucesso
        '''
        if(not isinstance(conta_destino, ContaCorrente)):
            raise ValueError('É necessário passar uma conta corrente para transferir')
            
        try:
            self.sacar(valor)
        except Exception as erro:
            print(erro)
            return False      

        conta_destino.depositar(valor)
        return True
    
class ContaVip(ContaCorrente):
    '''
    Cria uma representação de uma conta vip
    
    Atributos
    ----------
    cliente : Cliente (herdado)
        Guarda o cliente da conta
    saldo : float (herdado)
        Saldo da conta corrente
    cheque_especial: float
        Guarda o limite do chque especial
    '''
    
    def __init__(self, cliente, cheque_especial=0.0):
        '''
        Construtor

        Parâmetros
        ----------
        cliente : Cliente
            Recebe um cliente tornando proprietário da conta vip
        cheque_especial: float (opcional)
            Recebe um valor de cheque especial
        '''
        if(not isinstance(cliente, Cliente)):
            raise ValueError('É necessário passar um cliente para a conta')
         
        super().__init__(cliente)
        self.cheque_especial = cheque_especial
         
    def sacar(self, valor):
        '''
        Saca um valor passado como parâmetro da conta do usuário,
        alterando o saldo da conta.
        Caso o parametro não seja passado corretamente, 
        ou caso não for possível realizar o saque por falta de saldo.
        O sistema levanta uma exceção

        Parâmetros
        ----------
        valor : float
            Recebe o valor a ser sacado
        
        Retornos
        -------
        bool:
            Retorna se a operação foi realizada com sucesso
        '''
        if(not (isinstance(valor, float) or isinstance(valor, int))):
            raise ValueError('O valor precisa ser númerico')
            
        if(valor <= 0):
            raise ValueError('O valor precisa ser maior que zero')
            
        if( (self.saldo + self.cheque_especial - valor) < 0):
            raise ValueError('A conta não possui saldo suficiente para esta operação')
            
        self.saldo -= valor
        return True
        
        
cliente1 = Cliente("Rodrigo", 33, "")
conta1 = ContaVip(cliente1, 500.0)
conta1.depositar(200)
print(conta1)
conta1.depositar(200)
print(conta1)
conta1.sacar(200)
print(conta1)
# Erro de saldo
#conta1.sacar(300)
print(conta1)

cliente2 = Cliente("Maria", 20, "")
conta2 = ContaCorrente(cliente2)
print(conta2)

conta1.transferir(conta2, 100)
print(conta1)
print(conta2)

conta1.transferir(conta2, 100)
print(conta1)
print(conta2)

conta1.transferir(conta2, 500.01)
print(conta1)
print(conta2)

conta2.transferir(conta1, 150.99)
print(conta1)
print(conta2)
```

15) Crie uma classe Quadrado, filha da classe Retângulo do exercício 2.

```python
class Retangulo:
    
    def __init__(self, lado_a, lado_b):
        self.lado_a = lado_a
        self.lado_b = lado_b
        
    def area(self):
        return self.lado_a * self.lado_b
    
    def __repr__(self):
        return f"Retangulo de lado a = {self.lado_a} e lado b = {self.lado_b}"
    
class Quadrado(Retangulo):
    '''
    Cria uma representação de um quadrado 
    '''
    
    def __init__(self, lado):
        '''
        Construtor

        Parâmetros
        ----------
        lado : float
            Recebe o valor do lado do quadrado
        '''
        super().__init__(lado, lado)
    
    def __repr__(self):
        return f"Quadrado de lado = {self.lado_a}"

quadrado = Quadrado(4)
print(quadrado.area())
quadrado
```
